```python
# Start Nubra Proxy in background
import threading, time
import uvicorn

def run_nubra():
    uvicorn.run('nubra_backend:app', host='0.0.0.0', port=9000, log_level='info')

def run_orch():
    uvicorn.run('orchestrator:app', host='0.0.0.0', port=9010, log_level='info')

t1 = threading.Thread(target=run_nubra, daemon=True)
t2 = threading.Thread(target=run_orch, daemon=True)
t1.start()
t2.start()
time.sleep(3)
print('✅ Nubra Proxy on :9000')
print('✅ Orchestrator on :9010')
```

```python
# Setup ngrok tunnels to get public URLs
from pyngrok import ngrok
import getpass

print('Get your auth token from https://dashboard.ngrok.com/get-started/your-authtoken')
token = getpass.getpass('ngrok auth token: ')
ngrok.set_auth_token(token)

tunnel1 = ngrok.connect(9000, 'http')
tunnel2 = ngrok.connect(9010, 'http')

print(f'\n🔗 Nubra Proxy: {tunnel1.public_url}')
print(f'🔗 Orchestrator: {tunnel2.public_url}')
print(f'\n📝 Add these to Vercel env vars:')
print(f'  VITE_API_BASE={tunnel1.public_url}')
print(f'  VITE_ORCH_BASE={tunnel2.public_url}')
```
